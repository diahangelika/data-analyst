```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objs as go
```

```python
df['country'] = df['country'].fillna('Unknown')

df['country'] = df['country'].str.strip()

df_exploded = df.assign(country=df['country'].str.split(',')).explode('country')

df_exploded = df_exploded[df_exploded['country'] != '']

df_exploded['country'] = df_exploded['country'].str.strip()

film_titles_per_country = df_exploded.groupby('country')['title'].count()

film_titles_per_country_sorted = film_titles_per_country.sort_values(ascending=False)

top_10 = film_titles_per_country_sorted.head(11)

top_10_pd = pd.DataFrame(top_10)

top_10_pd

```

```python
import geopandas as gpd
import folium

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

world = world.merge(pd.DataFrame(film_titles_per_country), how='left', left_on='name', right_index=True)

world['title'].fillna(0, inplace=True)

m = folium.Map(location=[0, 0], zoom_start=2)

folium.Choropleth(
    geo_data=world,
    name='choropleth',
    data=world,
    columns=['name', 'title'],
    key_on='feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Number of Film Titles'
).add_to(m)

# tooltips
for idx, row in world.iterrows():
    folium.Marker(
        location=[row.geometry.centroid.y, row.geometry.centroid.x],
        tooltip=f"{row['name']} ({row['title']} films)"
    ).add_to(m)

# Display map
m.save('map.html')
m
```

```python
import geopandas as gpd

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

world = world.merge(pd.DataFrame(film_titles_per_country), how='left', left_on='name', right_index=True)

#world['title'].fillna(0, inplace=True)

# Plot choropleth map
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
world.plot(column='title', cmap='viridis', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True,
           legend_kwds={'label': 'Number of Film Titles'})
ax.set_title('Number of Film Titles per Country')
plt.show()
```

```python
tv_shows_df = df[(df['type'] == 'TV Show') & (df['duration'].str.contains(r'\d+ Season[s]?'))]
tv_shows_df['Seasons'] = tv_shows_df['duration'].str.extract(r'(\d+) Season[s]?').astype(int)

sorted_df = tv_shows_df.sort_values(by='Seasons', ascending=False)

top_10_tv_shows = sorted_df.head(10)

# Print the top 10 TV shows
# print(top_10_tv_shows[['title', 'Seasons']])

top_shows = pd.DataFrame(top_10_tv_shows[['title', 'Seasons']])

top_shows.plot.bar(x='title', y='Seasons', color="navy")
top_shows
```

```python
genres = df['listed_in'].str.split(', ').explode()

genre_counts = genres.value_counts()


top_10_genres = genre_counts.head(10)

top_gen = pd.DataFrame(top_10_genres)

top_gen.plot.bar(color = "green")
top_gen
```

```python
direct = df['director'].str.split(', ').explode()


director_counts = direct.value_counts()

top_10_director = director_counts.head(10)

top_dir = pd.DataFrame(top_10_director)

top_dir.plot.bar( color = "blue")
top_dir
```

```python
cast = df['cast'].str.split(', ').explode()

cast_counts = cast.value_counts()

top_10_cast = cast_counts.head(10)

top_cas = pd.DataFrame(top_10_cast)

top_cas.plot.bar(color = "maroon")
plt.legend()
top_cas
```
