```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv('/content/drive/MyDrive/BOOTCAMP/CLEANING & VISUALIZATION/Cause-of-Death-Indonesia.csv')
```

```python
data
```

```python
# Grafik garis (plot) 1 garis

group = data[['Year', 'Total Deaths']]
group_data = pd.DataFrame(group.groupby('Year')['Total Deaths'].sum().reset_index())

plt.figure(figsize=(15,5))
plt.plot(group_data['Year'], group_data['Total Deaths'])
plt.xlabel('Tahun')
plt.ylabel('Total Deaths')
plt.title('Grafik Garis')

plt.xticks(group_data['Year'], rotation=45)
plt.show()
```

```python
# Grafik garis (plot) 2 garis

group = data[['Year', 'Total Deaths', 'Data Redundancy']]
group_data = pd.DataFrame(group.groupby('Year')['Total Deaths','Data Redundancy'].sum().reset_index())

plt.figure(figsize=(15,5))
plt.plot(group_data['Year'], group_data['Total Deaths'], label='Garis 1', color='blue', linestyle=":",linewidth=2)
plt.plot(group_data['Year'], group_data['Data Redundancy'], label='Garis 2', color='green', linestyle="--",linewidth=2)
plt.xlabel('Tahun')
plt.ylabel('Total Deaths and redudancy')
plt.title('Grafik Garis')
plt.legend()

plt.xticks(group_data['Year'], rotation=45)
plt.show()
```

```python
# Grafik persebaran(scatter) 1 titik

group = data[['Year', 'Total Deaths']]
group_data = pd.DataFrame(group.groupby('Year')['Total Deaths'].sum().reset_index())

plt.figure(figsize=(15,5))
plt.scatter(group_data['Year'], group_data['Total Deaths'])
plt.xlabel('Tahun')
plt.ylabel('Total Deaths')
plt.title('Grafik Garis')

plt.xticks(group_data['Year'], rotation=45)
plt.show()
```

```python
# Grafik persebaran(scatter) 2 titik

group = data[['Year', 'Total Deaths', 'Data Redundancy']]
group_data = pd.DataFrame(group.groupby('Year')['Total Deaths','Data Redundancy'].sum().reset_index())

plt.figure(figsize=(15,5))
plt.scatter(group_data['Year'], group_data['Total Deaths'], label='Garis 1', color='blue')
plt.scatter(group_data['Year'], group_data['Data Redundancy'], label='Garis 2', color='green')
plt.xlabel('Tahun')
plt.ylabel('Total Deaths and redudancy')
plt.title('Grafik Garis')
plt.legend()

plt.xticks(group_data['Year'], rotation=45)
plt.show()
```

```python
# grafik histogram (hist) ga boleh pake data group by

plt.figure(figsize=(15,5))
plt.hist(data['Year'])

plt.xticks(data['Year'], rotation=45)
plt.show()
```

```python
# grafik histogram (hist) ga boleh pake data group by

plt.figure(figsize=(15,5))
plt.hist(data['Total Deaths'])

plt.show()
```

```python
# grafik histogram (hist) ga boleh pake data group by

plt.figure(figsize=(15,5))
plt.hist(data['Year'], alpha = 0.3, bins = 30, color = "red")

plt.xticks(data['Year'], rotation=45)
plt.show()
```

```python
# grafik boxplot (hist) ga bisa pake xsticks

plt.figure(figsize=(15,5))
plt.boxplot(data['Total Deaths'])

plt.show()
```

```python
for key,value in data['Total Deaths'].items():
  if value > 12000:
    print(value)
```

```python
key_outliar = []

for key,value in data['Total Deaths'].items():
  if value > 12000:
    key_outliar.append(key)

for i in key_outliar:
  print(data.iloc[i, 4:5])
```

```python
# grafik barplot (bar)

group = data[['Year', 'Total Deaths', 'Data Redundancy']]
group_data = pd.DataFrame(group.groupby('Year')['Total Deaths','Data Redundancy'].sum().reset_index())

width = 0.4
plt.figure(figsize=(15,5))
plt.bar(group_data['Year']-width/2, group_data.iloc[0]['Total Deaths'],label = "Total Deaths", alpha = 1, width = width)
plt.bar(group_data['Year']+width/2, group_data.iloc[0]['Data Redundancy'], label = "Redudancy", alpha = 1, width=width)
plt.xlabel('Tahun')
plt.ylabel('Total Deaths and redudancy')

plt.xticks(group_data['Year'])
plt.legend()
```

```python
# Grafik persebaran(scatter) 2 titik

group = data[['Year', 'Total Deaths', 'Data Redundancy']]
group_data = pd.DataFrame(group.groupby('Year')['Total Deaths','Data Redundancy'].sum().reset_index())

plt.figure(figsize=(15,5))
plt.scatter(group_data['Year'], group_data['Total Deaths'], label='Garis 1', color='maroon', alpha = 0.4, s = group_data['Total Deaths']/500)
plt.scatter(group_data['Year'], group_data['Data Redundancy'], label='Garis 2', color='green', alpha = 0.4, s = group_data['Data Redundancy']/500)
plt.xlabel('Tahun')
plt.ylabel('Total Deaths and redudancy')
plt.title('Grafik Garis')
plt.legend()

plt.xticks(group_data['Year'], rotation=45)
plt.show()
```

```python
# grafik barplot (bar) 2

grup = toko[['Ship Mode', 'Quantity']]
grup_toko = pd.DataFrame(grup.groupby('Ship Mode')['Quantity'].sum().reset_index())

plt.figure(figsize=(15,5))
plt.bar(grup_toko['Ship Mode'], grup_toko['Quantity'], color ='maroon', width = 0.4)

plt.show()
```

```python
# grafik barplot (bar) 3

grup = toko[['Region', 'Sales']]
grup_toko = pd.DataFrame(grup.groupby('Region')['Sales'].sum().reset_index())

plt.figure(figsize=(15,5))
plt.bar(grup_toko['Region'], grup_toko['Sales'], color ='maroon', width = 0.4)

plt.show()
```

```python
# grafik garis 4

grup = toko[['Category', 'Profit', 'Sales']]
grup_data = pd.DataFrame(grup.groupby('Category')['Profit','Sales'].sum().reset_index())


plt.figure(figsize=(15,5))
plt.scatter(grup_data['Category'], grup_data['Profit'],label='Profit', color='maroon')
plt.scatter(grup_data['Category'], grup_data['Sales'], label='Sales', color='blue')
plt.xlabel('Category')
plt.ylabel('Profit and Sales')

# plt.xticks(group_data['Year'])
plt.legend()
```

```python
# Grafik persebaran(scatter) 5

grup = toko[['Discount', 'Profit']]
grup_toko = pd.DataFrame(grup.groupby('Discount')['Profit'].sum().reset_index())

plt.figure(figsize=(15,5))
plt.scatter(grup_toko['Discount'], grup_toko['Profit'])
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.title('Grafik Hubungan Diskon dan Profit')

plt.xticks(grup_toko['Discount'], rotation=45)
plt.show()
```
