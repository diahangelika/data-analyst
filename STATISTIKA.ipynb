```python
#cara 1

data_bap = data_fr["bap_tilang"]

data_bap
```

```python
mean_bap = data_bap.mean()
mean_bap
```

```python
median_bap = data_bap.median()
median_bap
```

```python
from statistics import mode

modus_bap = mode(data_bap)

modus_bap
```

```python
varian_bap = data_bap.var()

varian_bap
```

```python
standar_deviasi_bap = data_bap.std()

standar_deviasi_bap
```

```python
# kuartil

q1 = data_bap.quantile(0.25) #angka-angkanya 0.25 0.5 0.75 ini udah fix rumusnya buat nentuin q1,q2,q3
q2 = data_bap.quantile(0.5)
q3 = data_bap.quantile(0.75)

print(f" q1 = {q1}")
print(f" q2 = {q2}")
print(f" q3 = {q3}")

#IQR
iqr = q3-q1

# Mencari data outlier
lower_bound = q1 - 1.5 * iqr
# lower_bound = max(q1 - 1.5 * iqr, min(data))
upper_bound = q3 + 1.5 * iqr

outliers = []

for x in data_bap:
  if x < lower_bound or x > upper_bound:
    outliers.append(x)

data_bersih = []

for x in data_bap:
  if x not in outliers:
    data_bersih.append(x)

print("Q1:", q1)
print("Q2 (Median):", q2)
print("Q3:", q3)
print("IQR:", iqr)
print("batas bawah:", lower_bound)
print("batas atas:", upper_bound)
print("outliers:",outliers)
print("data bersih:",data_bersih)
```

```python
bulan = data_fr["bulan"]
a = "" # kalo pake list ([]) harus diakses pake index
rata_rata = {} # set dictionary tidak bisa data ganda makannya yang tampil mau dia satu aja datanya
for i in bulan:
  if i !=a:
    a = i
    hasil = data_fr.loc[data_fr["bulan"] == i, "stop_operasi"].mean()
    rata_rata.update({i : hasil}) # set dia pake update nambahin datanya

print(rata_rata)
```
