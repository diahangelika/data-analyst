```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/BOOTCAMP/CLEANING & VISUALIZATION/dataset_final_project.csv')
```

```python
df
```

# 1

```python
df['Branch'].replace('A', 1, inplace = True)
df['Branch'].replace('B', 2, inplace = True)
df['Branch'].replace('C', 3, inplace = True)

df
```

```python
df['City'].replace('Yangon', 1, inplace = True)
df['City'].replace('Naypyitaw', 2, inplace = True)
df['City'].replace('Mandalay', 3, inplace = True)

df
```

```python
df['Customer type'].replace('Normal', 1, inplace = True)
df['Customer type'].replace('Member', 2, inplace = True)

df
```

```python
df['Product line'].replace('Electronic accessories', 1, inplace = True)
df['Product line'].replace('Fashion accessories', 2, inplace = True)
df['Product line'].replace('Food and beverages', 3, inplace = True)
df['Product line'].replace('Health and beauty', 4, inplace = True)
df['Product line'].replace('Home and lifestyle',5, inplace = True)
df['Product line'].replace('Sports and travel', 6, inplace = True)

df
```

```python
df['Payment'].replace('Cash', 1, inplace = True)
df['Payment'].replace('Credit card', 2, inplace = True)
df['Payment'].replace('Ewallet', 3, inplace = True)

df
```

# 2

```python
df['Total'] = df['Unit price'] * df['Quantity']
df['Tax'] = df['Total'] * 5/100
df['Final_Price'] = df['Total'] - df['Tax']
df
```

# 3

```python
df[df.duplicated()]
```

```python
df.fillna(0, inplace=True)
df
```

```python
#A
grup = df[['Product line', 'Quantity']]
grup_df = pd.DataFrame(grup.groupby('Product line')['Quantity'].sum().reset_index())

#plt.figure(figsize=(15,5))
plt.bar(grup_df['Product line'], grup_df['Quantity'], label = 'Quantity', width = 0.4)

plt.legend()
```

```python
# B 1

cus_norm = []
for i, y in df['Customer type'].items():
    if y == 1:
        cus_norm.append(1)

filtered_df = df[df['Customer type'] == 1]

grup_df = pd.DataFrame({'Product line': filtered_df['Product line'], 'Customer type': cus_norm})
grup_df = grup_df.groupby('Product line')['Customer type'].sum().reset_index()

# Plot the bar chart
plt.figure(figsize=(15, 5))
plt.bar(grup_df['Product line'], grup_df['Customer type'], label='Customer type', width=0.4)
plt.legend()
plt.show()

```

```python
# B 2

cus_member = []
for i, y in df['Customer type'].items():
    if y == 2:
        cus_member.append(2)

filtered_df = df[df['Customer type'] == 2]

grup_df = pd.DataFrame({'Product line': filtered_df['Product line'], 'Customer type': cus_member})
grup_df = grup_df.groupby('Product line')['Customer type'].sum().reset_index()

# Plot the bar chart
plt.figure(figsize=(15, 5))
plt.bar(grup_df['Product line'], grup_df['Customer type'], label='Customer type', width=0.4)
plt.legend()
plt.show()

```

```python
# C

grup = df[['Payment', 'Invoice ID']]
grup_df = pd.DataFrame(grup.groupby('Payment')['Invoice ID'].count().reset_index())

plt.figure(figsize=(10,5))
plt.bar(grup_df['Payment'], grup_df['Invoice ID'], label = 'Invoice ID', width = 0.4)

plt.legend()
plt.show()
```
