```python
mati = pd.read_csv('/content/drive/MyDrive/BOOTCAMP/CLEANING & VISUALIZATION/Cause-of-Death-Indonesia.csv')
```

```python
mati
```

```python
matibaru =mati[["Total Deaths","Year"]]

mdf = pd.DataFrame(matibaru.groupby("Year")["Total Deaths"].sum()) # yang kiri ("year") ini itu kategorinya, yang kanan ["total deaths"] ini datanya

mdf
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(13, 5))
sns.barplot(x = "Year", y = "Total Deaths", data = mdf)
```

```python
mdf.idxmax() # ini nyari total mati yang paling banyak tahun berapa pake idxmax(), soalnya tahunnya kan jadi id di data frame mdf
```

```python
bencana_sosial = mati[(mati['Type'] == "Bencana Sosial")]
bencana_sosial
```

```python
filter = bencana_sosial[['Total Deaths', 'Cause']]

filter_gabung = pd.DataFrame(filter.groupby('Cause')['Total Deaths'].sum())

filter_gabung
```

```python
sns.barplot(x="Total Deaths", y = "Cause", data = filter_gabung)
```

```python
filter_gabung.idxmax()
```

```python
filter_gabung.idxmin()
```

**NYOBAK**

```python
bencana_non_alam = mati[(mati["Type"] == "Bencana Non Alam dan Penyakit")]

bencana_non_alam
```

```python
non_alam = bencana_non_alam[["Year","Data Redundancy"]]

non_alam2 = pd.DataFrame(non_alam.groupby('Year')['Data Redundancy'].sum())

non_alam2
```

```python
plt.figure(figsize=(13, 5))
sns.barplot(x='Year', y='Data Redundancy', data = non_alam2)
```

```python
non_alam2.idxmax()
```

```python
non_alam2.idxmin()
```
