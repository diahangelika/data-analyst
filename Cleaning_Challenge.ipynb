```python
import pandas as pd
```

```python
desf = pd.read_excel("/content/drive/MyDrive/BOOTCAMP/CLEANING & VISUALIZATION/CHALLENGES/penumpang-transjakarta-desember-2021-excel-3.xls")
novf = pd.read_csv("/content/drive/MyDrive/BOOTCAMP/CLEANING & VISUALIZATION/CHALLENGES/penumpang-transjakarta-november-2021-2.csv")
```

```python
desf
```

```python
desf = desf.drop(columns=[col for col in desf.columns if 'Unnamed' in col])
```

```python
desf
```

```python
# hapus "pimpinan" di kolom yang seharusnya tahun

idx_pimpinan = desf[desf['tahun'] == "pimpinan"].index[0]
desf.drop(idx_pimpinan, inplace=True)

desf
```

```python
#buang semicolon di nama kolom jumlah penumpang

desf.rename(columns= lambda x: x.strip(';'), inplace=True)

desf
```

```python
# hapus NaN, dropna ini menghapus SATU BARIS SEMUANYA yang ada nilai Nan nya

desf = desf.dropna()

desf
```

```python
# buang semicolon di value kolom jumlah penumpang

def remove_object(x):
  if isinstance(x, str):
    x = x.replace(';','')
  return x
```

```python
# applymap itu kaya looping

kolom = desf.columns
desf[kolom] = desf[kolom].applymap(str)
desf = desf.applymap(remove_object)

desf
```

```python
desf.info()
```

```python
desf.head(52)
```

```python
def to_float(x):
  try:
    return float(x)
  except ValueError:
    return None

desf['jumlah_penumpang'] = desf['jumlah_penumpang'].map(to_float)

desf.dropna()

desf
```

```python
desf = pd.concat([desf, novf], ignore_index = True)

desf
```

```python
desf.to_excel('/content/drive/MyDrive/BOOTCAMP/CLEANING & VISUALIZATION/CHALLENGES/data_bersih.xlsx', index=False)
```
