```python
total = df.iloc[0: , 2:]
tot = total.sum()

dt_tot = []
for i,y in tot.items():
  dt_tot.append(y)

total_turis = sum(dt_tot)

print(f"total turis pada tahun 2018 dan 2019 adalah {int(total_turis)}")
```

```python
rata_turis = total_turis / len(dt_tot)

print(f"rata-rata turis pada tahun 2018 dan 2019 adalah {int(rata_turis)}")
```

**TOTAL DAN RATA RATA TURIS PADA TAHUN 2018 DAN 2019**

```python
df_2019 = df.loc[2019]
df_2018 = df.loc[2018]

print("Data for 2019:")
dt2019 = df_2019.iloc[0: , 2:]
print(dt2019)


print("\nData for 2018:")
dt2018 = df_2018.iloc[0: , 2:]
print(dt2018)
```

```python
tot19 = dt2019.sum()
tot19
```

```python
tot19 = dt2019.sum()

tot_2019 = []
for i,y in tot19.items():
  tot_2019.append(y)

turis_2019 = sum(tot_2019)
rata_turis19 = turis_2019 / len(tot_2019)

print(f"total turis pada tahun 2019 adalah {int(turis_2019)}")
print(f"rata rata turis pada tahun 2019 adalah {int(rata_turis19)}")
```

```python
tot18 = dt2018.sum()

tot_2018 = []
for i,y in tot18.items():
  tot_2018.append(y)

turis_2018 = sum(tot_2018)
rata_turis18 = turis_2018 / len(tot_2018)

print(f"total turis pada tahun 2018 adalah {int(turis_2018)}")
print(f"rata rata turis pada tahun 2018 adalah {int(rata_turis18)}")
```

**TURIS PALING BANYAK PADA TAHUN 2018 DAN 2019**

```python
max_month_2019 = tot19.idxmax()

max_value_2019 = tot19[max_month_2019]

print(f"total turis paling banyak tahun 2019 adalah bulan {max_month_2019} dengan total {int(max_value_2019)}")
```

```python
max_month_2018 = tot18.idxmax()

max_value_2018 = tot18[max_month_2018]

print(f"total turis paling banyak tahun 2019 adalah bulan {max_month_2018} dengan total {int(max_value_2018)}")
```

**TURIS PALING SEDIKIT PADA TAHUN 2018 DAN 2019**

```python
min_month_2019 = tot19.idxmin()

min_value_2019 = tot19[min_month_2019]

print(f"total turis paling sedikit tahun 2019 adalah bulan {min_month_2019} dengan total {int(min_value_2019)}")
```

```python
min_month_2018 = tot18.idxmin()

min_value_2018 = tot18[min_month_2018]

print(f"total turis paling sedikit tahun 2018 adalah bulan {min_month_2018} dengan total {int(min_value_2018)}")
```

**DATA FRAME BARU**

```python
# Membuat DataFrame

mon = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
data_turis = {
    'Year': [2018] * 12 + [2019] * 12,
    'Bulan': mon*2,
    'Total': tot_2018 + tot_2019,
}

df_turis = pd.DataFrame(data_turis)

#df_turis

# Memastikan semua bulan ada di DataFrame
all_months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df_turis['Bulan'] = pd.Categorical(df_turis['Bulan'], categories=all_months, ordered=True)

df_turis_pivot = df_turis.pivot_table(index='Year', columns='Bulan', values='Total', aggfunc='sum')

#df_turis_pivot

# Menambahkan kolom total kedatangan dalam setahun dan rata-rata kedatangan dalam setahun
df_turis_pivot['Max'] = [max_month_2018, max_month_2019]
df_turis_pivot['Min'] = [min_month_2018, min_month_2019]
df_turis_pivot['total kedatangan dalam setahun'] = [turis_2018, turis_2019]
df_turis_pivot['Rata-rata kedatangan dalam setahun'] = [rata_turis18, rata_turis19]

df_turis2 = pd.DataFrame(df_turis_pivot)
df_turis2
```
